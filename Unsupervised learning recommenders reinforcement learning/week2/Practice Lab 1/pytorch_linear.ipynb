```python
# 用autograd拟合一条直线

import torch
```

```python
x = torch.arange(0, 10, 1)
print(x.shape)

y = 2 * x + 1
print(y.shape)
```

```python
# 为什么randn就不对
# rand randn有什么区别
# randn是标准正态分布
# rand是0-1均匀分布
w = torch.rand(1, requires_grad=True)
b = torch.rand(1, requires_grad=True)

# y_hat = w * x + b
# print(y_hat.shape)
# 难道是参数的形状的问题

optimizer = torch.optim.SGD([w, b], lr=0.01)

epoch = 100
for e in range(epoch):

    # prediction 
    # y_hat = w * x + b
    # compute loss
    loss = torch.mean((w * x + b - y) ** 2)
    print(loss.item())

    # backward
    loss.backward()

    # gradient descent
    optimizer.step()
    # 不调用zero确实会出问题
    optimizer.zero_grad()

print(w, b)
```
